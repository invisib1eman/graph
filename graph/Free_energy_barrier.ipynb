```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
V=[1,2,3,4]
F=[3,6,9,-6]
```

```python
fig=plt.figure()
ax=fig.add_subplot()
ax.scatter(V,F)
ax.plot(V,F)
ax.set_xlabel("Vertices",fontsize=20)
ax.set_ylabel("F/kT",fontsize=20)
plt.xticks(np.arange(1,5,1))
plt.savefig("Free_energy.png")
```
